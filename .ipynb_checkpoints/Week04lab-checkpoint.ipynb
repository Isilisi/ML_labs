```python
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

## Question 1

```python
iris = datasets.load_iris()
print(iris.DESCR)
```

```python
X = iris.data
y = iris.target
display(np.shape(X))
display(y.shape)
```

```python
dir(iris)
```

## Question 2

```python
from sklearn.dummy import DummyClassifier
zero_r = DummyClassifier(strategy = 'most_frequent')
```

```python
zero_r.fit(X, y)
yhat = zero_r.predict(X)
display(yhat)
zero_r.score(X, y)
```

```python
display(Counter(y))
Counter(yhat)
```

```python
# weighted random classifier constructs a probability distribution over the labels in the training data
# then classifies by sampling from that probability distribution

weighted_clf = DummyClassifier(strategy='stratified')
weighted_clf.fit(X, y)

accuracies = [weighted_clf.score(X, y) for i in range(100)]
display(np.mean(accuracies))
plt.plot(range(100), accuracies)
plt.hlines(np.mean(accuracies), xmin=0, xmax=100, color='red')
plt.show()
```

Yes, on average, it has the same accuracy

## Question 3

```python
from sklearn.tree import DecisionTreeClassifier

# 1-R classifier is the same as Decison Tree with a single partition
one_r = DecisionTreeClassifier(max_depth=1)
one_r.fit(X, y)
display(one_r.score(X, y))

# Decision Tree classifier
dt = DecisionTreeClassifier(max_depth=3)
dt.fit(X, y)
display(dt.score(X, y))
```

```python
display(one_r.feature_importances_)
display(dt.feature_importances_)
```

```python
#predictions = one_r.predict(X) # one R uses the last column (shown as blue on the plot)
predictions = dt.predict(X) # the decision tree uses green and blue points

plt.scatter(X[:, 1], predictions+0.02, color='red', alpha=0.6)
plt.scatter(X[:, 2], predictions, color='green', alpha=0.6)
plt.scatter(X[:, 3], predictions, color='blue', alpha=0.6)
plt.show()
```

```python
# Information gain and Gini coefficient give the same results most of the time
```

```python
# a better way to visualise a decision tree

from sklearn.externals.six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus

dot_data = StringIO()

export_graphviz(dt, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

## Question 4

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

X_train, X_test, y_train, y_test = train_test_split(X, y)

zero_r.fit(X_train, y_train)
one_r.fit(X_train, y_train)
dt.fit(X_train, y_train)

for model in [zero_r, one_r, dt]:
    print(accuracy_score(model.predict(X_train), y_train), accuracy_score(model.predict(X_test), y_test))
```

```python
test_accuracies = []
for proportion in np.arange(0.01, 1, 0.01):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=proportion)
    dt.fit(X_train, y_train)
    test_accuracies.append(accuracy_score(dt.predict(X_test), y_test))
    
plt.plot(np.arange(0.01, 1, 0.01), test_accuracies)
plt.show()
```

## Question 5

```python
from sklearn.model_selection import cross_val_score

for model in [zero_r, one_r, dt]:
    display(cross_val_score(model, X, y, cv=10))
```
